```python
import os
from dotenv import load_dotenv
```

```python
load_dotenv()
```

```python
if not os.getenv("GROQ_API_KEY"):
    raise ValueError("Groq api key not found in .env")
else:
    print("groq api key found in .env")
```

```python
if not os.getenv("GOOGLE_API_KEY"):
    raise ValueError("google api key not found in .env")
else:
    print("google api key found in .env")
```

```python
from mem0 import Memory
config = {
    "llm": {
        "provider": "groq",
        "config": {
            "model": "llama-3.3-70b-versatile",
            "api_key": os.environ ["GROQ_API_KEY"],
            "temperature": 0,
        },
    },
    "embedder": {
        "provider": "gemini",
        "config": {
            "model": "models/gemini-embedding-001",
            "embedding_dims": 1536,
            "api_key": os.environ["GOOGLE_API_KEY"],
        }

    },

    "vector_store": {
        "provider": "qdrant",
        "config": {
            "path": ":memory:",
            "embedding_model_dims": 1536,
        }
    }
}
m=Memory.from_config(config)
```

```python
messages = [
    {"role": "user", "content": "Hi, I'm Deep. I love basketball and gaming."},
    {"role": "assistant", "content": "Hey Deep! I'll remember your interests."}
]
m.add(messages, user_id="deep")
```

```python
results = m.search("What do you know about me?", filters={"user_id": "deep"})
print(results)
```

```python
messages = [
    {"role": "user", "content": "I'm a computer science student studying at Institute of Engineering, Thapathali Campus."},
    {"role": "assistant", "content": "I'll remember your academic background."}
]
m.add(messages, user_id="deep")
```

```python
results = m.search("What do you know about me?", filters={"user_id": "deep"})
results=results['results']

for i in range(len(results)):
    print(results[i]['memory'])
```
